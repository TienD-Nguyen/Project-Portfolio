```python
def filling_missing_director(dataset):
    for instance in range(len(dataset)):
        if dataset["director"].isnull()[instance]:
            title_list = ia.search_movie(dataset["title"][instance])
            
            if title_list:
                title_id = title_list[0].movieID
                movie = ia.get_movie(title_id)
                try:
                    director = movie["directors"][0]["name"]
                except KeyError:
                    director = "NULL"
            else:
                director = "NULL"
                
            dataset["director"][instance] = director
    return dataset
```

```python
def filling_missing_country(dataset):
    for instance in range(len(dataset)):
        if dataset["country"].isnull()[instance]:
            title_list = ia.search_movie(dataset["title"][instance])
            
            if title_list:
                title_id = title_list[0].movieID
                movie = ia.get_movie(title_id)
                try:
                    country = movie["countries"][0]
                except KeyError:
                    country = "NULL"
            else:
                country = "NULL"
                
            dataset["country"] = country
    return dataset
```

```python
netflix = filling_missing_director(netflix)
netflix = filling_missing_country(netflix)
```

```python
for title in netflix.title:
    title_list = ia.search_movie(title)
    try:
        type_id = title_list[0].movieID
    except IndexError:
        print(title)
```
