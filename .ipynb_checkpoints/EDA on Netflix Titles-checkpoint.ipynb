### 3. Data Cleaning, Wrangling and Understanding

As there some NaN values can be found through intial exploration of the dataset in the above section. Therefore, examining missing values is required before continuing with analysis

```python
# Examining which columns contain null values
netflix.isnull().sum()
```

```python
# Calculating the percent of missing values in its associated columns
for i in netflix.columns:
    null_rate = netflix[i].isnull().sum() / len(netflix) * 100
    if null_rate > 0:
        print(f"{i}'s null rate: {null_rate:.2f}%")
```

From the exploration on the total of missing values in the dataset, it can be observed that is required extreme treatment before conducting an analysis on the dataset.
* With the "director" and "cast" features on the dataset, they would not contribute much on the analysis and highlight any trend or insights to persuade a famous actor/actress or director since some movies with famous casts often associate with their production company and not appear on Netflix sometimes.
* "rating" and "duration" would be handled with manually scrapping the web for movie's rates and filling out common movies/tv shows duration for the attribute.
* "data_added" would not be treated specially
* "country' would be filled through scrapping the IMBD database.

Let's proceed to removing and handling columns that contain missing data

```python
# Let's examing unique movies' ratings first.
netflix["rating"].unique()
```

```python
# Checking instances that contain missing values in "rating" feature.
netflix[netflix.rating.isna()]
```

```python
# There are those titles with wrongly entered rating values. Let's have a look on them.
error_instances = pd.DataFrame(columns=netflix.columns)
for i in ['74 min', '66 min', '84 min']:
    error_instances = error_instances.append(netflix[netflix.rating == i])
error_instances
```

Based on information found on IMDB rating information and Netflix platforms, those values can be manually replaced.

```python
# Create a dictionary contain instances misses values on rating and wrongly entered.
rating_replacement = {
    5541: "TV-MA",
    5813: "TV-MA",
    5794: "TV-MA",
    5989: "PG",
    6827: "TV-PG",
    7312: "TV-G",
    7537: "PG-13"
}

# for-loop to replace current rating values with declared.
for id, rate in rating_replacement.items():
    netflix.iloc[id, 8] = rate

# Re-check to see if there is still missing values.
print(f"Unique rating: {netflix.rating.unique()}")
print(f"Total of missing values in 'rating': {netflix.rating.isnull().sum()}")
```

```python
# Dropping "director" and "cast" from the dataset as mentioned above.
netflix = netflix.drop(["director", "cast"], axis=1)
netflix.columns
```

As mentioned, the "country" features might be useful for the analysis as it helps determine which country produce most movie titles on Netflix and content from which country is often added to Netflix

Therefore, in the next cell, IMBb API will be utilised to scrapping information from IMDB databased and fill-up missing values in "country" features.

```python
# Import the IMBD library to scrapping the movie's production country.
from imdb import IMDb
ia = IMDb()

# Defining function that will take the name of movie title and comparing with IMDB database to return associated Movie ID.
# The Movie ID would then be used to retrieve other information, in this case, production country information.
def filling_missing_country(dataset):
    for instance in range(len(dataset)):
        if dataset["country"].isnull()[instance]:
            title_list = ia.search_movie(dataset["title"][instance])
            
            if title_list:
                title_id = title_list[0].movieID
                movie = ia.get_movie(title_id)
                # "try" statement would help ignore records do not contain "country" information or not recorded.
                try:
                    country = movie["countries"][0]
                    dataset["country"][instance] = country
                except KeyError:
                    pass
    return dataset

netflix = filling_missing_country(netflix)
netflix["country"].isnull().sum()
```

After scrapping the IMBD database, there are still 305 instances that contains missing values on "country" attribute. As a result, "mode" function would be applied to fill-in with most common values in the features.

```python
netflix["country"] = netflix["country"].fillna(netflix["country"].mode()[0])

# Re-check to ensure there would no further instances in "country" column contains missing values.
netflix["country"].isnull().sum()
```

```python
# Filter out any instances that counting missing values on 'date_added' column
netflix = netflix[netflix["date_added"].notna()]
```

```python
# Splitting "year_added" column into different separated data columns for visualisation later.
netflix["year_added"] = netflix["date_added"].apply(lambda x: x.split(",")[-1])
netflix["year_added"] = netflix["year_added"].apply(int)
netflix["year_added"].head(5)
```

```python
# Splitting to attain "month" columns for visualisation and pattern finding latter.
netflix["month_added"] = netflix["date_added"].apply(lambda x : x.split(" ")[0])
netflix["month_added"].head(5)
```

```python
# Creating extra column that defind appropriate audience for each title's rating type.
rating_ages = {
    'TV-PG': 'Older Kids',
    'TV-MA': 'Adults',
    'TV-Y7-FV': 'Older Kids',
    'TV-Y7': 'Older Kids',
    'TV-14': 'Teens',
    'R': 'Adults',
    'TV-Y': 'Kids',
    'NR': 'Adults',
    'PG-13': 'Teens',
    'TV-G': 'Kids',
    'PG': 'Older Kids',
    'G': 'Kids',
    'UR': 'Adults',
    'NC-17': 'Adults'
}

# Mapping the rating_ages with titles' rates in the dataset and create new a new column.
netflix["target_ages"] = netflix["rating"].replace(rating_ages)
netflix["target_ages"].unique()
```

```python
# Formating the correct datatype for attributes in the dataset.
netflix["type"] = pd.Categorical(netflix["type"])
netflix["target_ages"] = pd.Categorical(netflix["target_ages"], categories=["Kids", "Older Kids", "Teens", "Adults"])
netflix.dtypes
```

As observed, there are few titles were produced in several countries at the same time. Therefore, these instances would be split out and map into a dictionary with their number of time appear on the dataset.

```python
# Using Counter function to counting number of time each country appear on the dataset after splitting to take unique country list
from collections import Counter
country_data = netflix["country"]
country_counting = pd.Series(dict(Counter(",".join(country_data).replace(" ,",",").replace(", ",",").split(",")))).sort_values(ascending=False)
country_counting
```

```python
# Using Pie plot to visualise top 20 producing contries on Netflix platform
viz_2 = px.pie(values=country_counting.values[:20],
               names=country_counting.keys()[:20],
               title="Distribution of Top 20 Producing Countries",)

viz_2.update_traces(hoverinfo="label+percent", textinfo="percent")
viz_2.show()
```

```python
# Using bar chart to visualise number of titles produced by each country
viz_3 = px.bar(y=country_counting.values[:20],
               x=country_counting.keys()[:20],
               color=country_counting.values[:20],
               labels={"x": "Country", "y": "Number of titles", "color": "No. Titles"})
viz_3.update_traces(textposition="outside", text=country_counting.values[:20], textfont_size=12,
                    hovertemplate="<br>".join(["Country: %{x}", "Number of Titles: %{y}"]))
viz_3.update_layout(title_text="Nunber of titles produced in top 20 associated countries")
viz_3.show()
```

From above two visuals, there is clear that United States produced most of the movie titles on Netflix, and followed by India and United Kingdom, etc. This details might indicate that film industry in United States is developed and expected by most audience around the globe.

Next, let's review number of title released by each year. However, the dataset contain values way back from 1925; therefore, we only consider instances from 1991.

```python
# Grouping the data based on released_year and type
released_year = netflix.loc[netflix["release_year"] >= 1991].groupby(["release_year", "type"]).agg({"show_id": "count"}).reset_index()

# Utilising Scatter plot with lines and markers to visualise number of titles released in each year from 1991.
viz_4 = go.Figure()
viz_4.add_trace(go.Scatter(x=released_year.loc[released_year["type"] == "TV Show"]["release_year"],
                           y=released_year.loc[released_year["type"] == "TV Show"]["show_id"],
                           mode="lines+markers",
                           name="TV Show: Released year",
                           marker_color="green"))

viz_4.add_trace(go.Scatter(x=released_year.loc[released_year["type"] == "Movie"]["release_year"],
                           y=released_year.loc[released_year["type"] == "Movie"]["show_id"],
                           mode="lines+markers",
                           name="Movie: Released year",
                           marker_color="red"))

viz_4.show()
```

From the above, the number of movie titles released in recent years had somehow decreased drastically while there is a slight increase in releasing TV Show titles. 

The opposite movement here might due to the fact that Netflix starts producing its own TV Shows and wanting to attract more audience on their TV Shows and generate profit from its TV Shows. From that assumptions, there is less movie titles appear on Netflix platform and lead to the decrease in number of titles based on its associated release years.

Also, let's us explore number of titles released based on each year and country that it's produced.

```python
# Grouping data based on release_year and production country
release_year_country = netflix.groupby("release_year")["country"].value_counts().reset_index(name="counts")

# Utilising Plotly's choropleth to visualise geographic data
fig = px.choropleth(release_year_country, locations="country", color="counts",
                    locationmode="country names",
                    animation_frame="release_year",
                    range_color=[0, 200],
                    color_continuous_scale=px.colors.sequential.OrRd,
                    hover_name="country",
                    hover_data={"country":False},
                    labels={"release_year": "Released Year", "counts": "No. Titles"})
fig.update_layout(title="Number of titles released by year in different countries")
fig.show()
```

Throughout almost a century, the U.S has always be the bigger producer of movie and tv shows that is now premiered on Netflix. There are only few years that the U.S did not produce any titles; otherwise, the U.S often contribute by little or much more values to the international film industry of the world.

```python
# Defining function that would group the data and counting number of appearance for titles' rating and target_ages
def generate_rating(dataset):
    movie_rating = netflix.groupby(["rating", "target_ages"]).agg({"show_id": "count"}).reset_index()
    movie_rating.columns = ["rating", "target_ages", "count"]
    movie_rating = movie_rating[movie_rating["count"] != 0]
    movie_rating = movie_rating.sort_values("target_ages")
    return movie_rating
```

```python
# Input original dataset to generate a new dataframe contains only titles' ratings with its associated target_ages
movie_rating = generate_rating(netflix)

# Using bar-chart to visualise each titles' rating and its associate number of title on Netflix.
viz_6 = px.bar(movie_rating, x="rating", y="count", color="target_ages",
               labels={"count": "Number of title", "rating": "Movie's Rating"})
viz_6.update_layout(title="Counting list of associated movie rating")
viz_6.show()
```

```python
# Split and join multiple movie genre and counting number of appearance in the dataset
def counting_genre(dataset):
    genre_data = dataset["listed_in"]
    genre_counting = pd.Series(dict(Counter(",".join(genre_data).replace(" ,",",").replace(", ",",").split(",")))).sort_values(ascending=False)
    return genre_counting

genre_counting = counting_genre(netflix)
genre_counting.index
```

```python
# Using bar chart to visualise top 20 popular genre on Netflix
viz_7 = px.bar(x=genre_counting.index[:20],
               y=genre_counting.values[:20],
               color=genre_counting.values[:20],
               labels={"x": "Genre", "y": "Number of Titles", "color": "No. Titles"})
viz_7.update_traces(textposition="outside", text=genre_counting.values[:20], textfont_size=12,
                    hovertemplate="<br>".join(["Titles' Genre: %{x}", "Number of Titles: %{y}"]))
viz_7.update_layout(title_text="Top 20 Genre listed on Netflix platform")
viz_7.show()
```

```python
# Create a dataframe that only contain Movie titles
movie_genre = counting_genre(netflix[netflix["type"] == "Movie"])

# Using bar chart to visualise top 20 popular genre of Movie type on Netflix
viz_8 = px.bar(x=movie_genre.index[:20],
               y=movie_genre.values[:20],
               color=movie_genre.values[:20],
               labels={"x": "Movies' Genre", "y":"Number of Title", "color":"No.Titles"})
viz_8.update_traces(textposition="outside", text=movie_genre.values[:20], textfont_size=12,
                    hovertemplate="<br>".join(["Movies' Genre: %{x}", "Number of Titles: %{y}"]))
viz_8.update_layout(title_text="Top 20 Movie Genre listed on Netflix platform")
viz_8.show()
```

```python
# Create a dataframe only contain TV Show titles
tvshow_genre = counting_genre(netflix[netflix["type"] == "TV Show"])

# Using bar chart to visualise top 20 popular genre of TV Show type on Netflix
viz_9 = px.bar(x=tvshow_genre.index[:20],
               y=tvshow_genre.values[:20],
               color=tvshow_genre.values[:20],
               labels={"x":"TV Shows' Genre", "y":"Number of Title", "color":"No. Titles"})
viz_9.update_traces(textposition="outside", text=tvshow_genre.values[:20], textfont_size=12,
                    hovertemplate="<br>".join(["TV Shows' Genre: %{x}", "Number of Titles: %{y}"]))
viz_9.update_layout(title_text="Top 20 TV Show Genre listed on Netflix platform")
viz_9.show()
```

For both Movie and TV Show, the most popular genres on Netflix service are International, Dramas and Comedies. These genres often watched by the audience and being produced more frequently.

```python
# Generate a dataframe that contain top 5 titles' genre for recent 5 years.
year_genre_df = netflix[(netflix["listed_in"].isin(genre_counting.index[:5])) & (netflix["year_added"] >= 2017)].groupby(["listed_in", "year_added"]).agg({"title": "count"})
year_genre_df = year_genre_df.reset_index()
year_genre_df.columns = ["genre" , "year_added", "count"]

# Using Sunburst to visualisee top 5 genre in recent 5 years
viz_10 = px.sunburst(year_genre_df, path=["year_added", "genre"], values="count")
viz_10.update_layout(height=800, width=800)
viz_10.show()
```

```python
# Generate a dataframe that contain only movie titles and to extract movies' duration information
movie_df = netflix[netflix["type"] == "Movie"]

# Due to a missing value was not handled, it would be filled in with the most common value in the columns
movie_df["duration"] = movie_df["duration"].fillna(movie_df["duration"].mode()[0])

# Apply lambda function to extract the duration number and convert it into string for comparation later
movie_duration = movie_df['duration'].apply(lambda x: int(x.split(" ")[0]))
```

```python
# Create a new conditional columns based on movie_duration.
movie_df.loc[movie_duration.loc[movie_duration < 90].index, "duration_bin"] = "Less than 1.30 hours"
movie_df.loc[movie_duration.loc[(movie_duration >= 90) & (movie_duration < 150)].index, "duration_bin"] = "More than 1.30 - less than 2.30 hours"
movie_df.loc[movie_duration.loc[movie_duration >= 150].index, "duration_bin"] = "More than 2.30 hours"

# Checking for changes if it's inplaced
movie_df
```

```python
# Using historgram to visualise the distribution of movies duration in the dataset
viz_11 = px.histogram(x=movie_df["duration_bin"],
                      color=movie_df["duration_bin"])
viz_11.update_xaxes(categoryorder="array",
                    categoryarray=["Less than 1.30 hours", "More than 1.30 - less than 2.30 hours", "More than 2.30 hours"])
viz_11.update_layout(title="Distribution of movies duration",
                     xaxis_title="Duration of the Movies")
viz_11.show()
```

From the above visualisation, it can be seen that high percent of movieo on Netflix platform often have more then 1.30 hours of watching time, which might highlights the common amount of time it take to deliver great director's story to the audience.

```python
# A data frame only contain TV Show titles
tvshow_df = netflix[netflix["type"] == "TV Show"]

# Using the histogram to visualise the distribution of TV Show duration.
viz_12 = go.Figure()
viz_12.add_trace(go.Histogram(x=tvshow_df["duration"], name="TV Shows' Seasons",
                              hovertemplate="<br>".join(["Number of Seasons: %{x}", "Total count: %{y}"]),))
viz_12.update_xaxes(categoryorder="total descending")
viz_12.update_layout(title="Distribution of TV Shows duration",
                     xaxis_title="Duration of TV Shows",
                     yaxis_title="Total count")
viz_12.show()
```

As it can be seen, most tv show series only have 1 season in total and the number of TV Show have more than 2 seasons drops dramtically after each other. However, there are maybe several reason that those TV Show only have 1 season in total:
* The number of episodes (often more than usual) they have in one season is enough to persuade and construct the intended story as a whole.
* The main story is still in the constructing stage, which is not finished and need additional seasons to build upon what happened and deliver the fullest of the story to audiences. This might result in lesser episodes in one seasons.
* Budget reason can be considered as the TV Show was over budget after the production of season 1, which might need to be stopped before produce it again for another season.
